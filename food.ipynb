```python
# handling data inconsistency
# converting strings into lowercase
df = df.apply(lambda x:x.str.lower() if x.dtype == "object" else x)
df.head()
```

```python
df.replace(['nun','none','null','nan','Unknown',], np.nan, inplace= True)
df.head()
```

```python
cleaned = df.dropna(axis=1 , thresh = 0.5)
cleaned.shape
```

```python
# filling the remaining missing values with the median 
num_col = cleaned.select_dtypes(include= ['int64','float64']).columns
cleaned[num_col] = cleaned[num_col].fillna(cleaned[num_col].median())
```

```python
cleaned.isnull().sum().sum()
```

```python
# filling categorical values with mod 
cat_col = cleaned.select_dtypes(include = ['object']).columns
cleaned[cat_col] = cleaned[cat_col].fillna(cleaned[cat_col].mode().iloc)
```

```python
cleaned.isnull().sum().sum()
```

```python
# checking dup
dup = cleaned.duplicated().sum()
dup
```

```python
cleaned = cleaned.drop_duplicates()
```

```python
col_convert = ['GPA',"weight"]
cleaned[col_convert] = cleaned[col_convert].apply(pd.to_numeric,errors = 'coerce')
```

```python
cleaned[col_convert].dtypes
```

```python
cleaned.isnull().sum().sum()
```

```python
remaining_missing = cleaned.isnull().sum()
remaining_missing
```

```python
# separating numerical and categorical missing values
numerical_missing = remaining_missing.index[cleaned[remaining_missing.index].dtypes != 'object']
categorical_missing = remaining_missing.index[cleaned[remaining_missing.index].dtypes == 'object']
```

```python
# filling numerical columns by mean
cleaned[numerical_missing] = cleaned[numerical_missing].fillna(cleaned[numerical_missing].mean())
```

```python
# filling categorical columns with mode
cleaned[categorical_missing] = cleaned[categorical_missing].fillna(cleaned[categorical_missing].mode().iloc[0])
```

```python
cleaned.isnull().sum().sum()
```

```python
# outliers
def outliers(df, cols):
    for x in cols:
        q1 = df[x].quantile(0.25)
        q3 = df[x].quantile(0.75)
        IQR = q3 - q1
        lower_bound = q1 - 1.5*IQR 
        upper_bound = q3 + 1.5*IQR 
        df[x] = df[x].clip(lower=lower_bound, upper=upper_bound)
    return df
cleaned = outliers(cleaned, num_col)
```
